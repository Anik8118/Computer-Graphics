```python
def Bresenham(x1, y1):
  x_point = []
  y_point = []
  del_x = x2 - x1
  del_y = y2 - y1
  p_not = 2*del_y - del_x
  for _ in range(del_x+1 if del_x > del_y else del_y+1):
    if p_not >= 0:
      x_point.append(x1)
      y_point.append(y1)
      x1+=1
      y1+=1
      p_not += 2*(del_y - del_x)

    else:
      x_point.append(x1)
      y_point.append(y1)
      x1+=1
      p_not += 2*del_y
  result = zip(x_point, y_point)
  for _ in result:
    print(_, end=" ")
  plt.plot(x_point, y_point, 'g.')
  plt.xlabel("X-axis")
  plt.ylabel("Y-axis")
  plt.title("Bresrham Line Drawing Algorithm")
```
