```python
import matplotlib.pyplot as plt
```

```python
def plot_circle(xc, yc, r):
    x = 0
    y = r
    p = 3 - 2 * r

    x_cord = []
    y_cord = []

    while x <= y:
        x_cord.extend([xc + x, xc - x, xc + x, xc - x, xc + y, xc - y, xc + y, xc - y])
        y_cord.extend([yc + y, yc + y, yc - y, yc - y, yc + x, yc + x, yc - x, yc - x])

        if p < 0:
            p = p + 4 * x + 6
        else:
            p = p + 4 * (x - y) + 10
            y -= 1

        x += 1
    plt.scatter(x_cord, y_cord, color='blue', s=10)
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
xc, yc, r = 0, 0, 10
plot_circle(xc, yc, r)
```
