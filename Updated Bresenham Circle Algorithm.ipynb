```python
import matplotlib.pyplot as plt
```

```python
p = 0
q = 0
r = int(input())
q = r
```

```python
p_not = 1-r
```

```python
x_cord = []
y_cord = []

x_cord.append(p)
y_cord.append(q)

while p < q:
    if p_not >= 0:
        p += 1
        q -= 1
        p_not = p_not + 4 * (p - q) + 10
    else:
        p += 1
        p_not = p_not + 4 * p + 6

    x_cord.append(p)
    y_cord.append(q)




```

```python
print(x_cord)
```

```python
print(y_cord)
```

```python
res = zip(x_cord, y_cord)
for _ in res:
  print(_, end=" ")
```

```python
for i in range(len(x_cord)):
  x = x_cord[i]
  y = y_cord[i]
  plt.scatter([x, -x, x, -x, y, -y, y, -y], [y, y, -y, -y, x, x, -x, -x], color='blue')
plt.title('Mid point Circle')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.savefig('Mid point cercle.png')
```
