```python
import matplotlib.pyplot as plt
```

```python
p = int(input())
q = int(input())
r = int(input())
q = r
```

```python
p_not = 1-r
```

```python
x_cord = []
y_cord = []
x_cord.append(p)
y_cord.append(q)

while p!=q:
  if p_not >= 0:
    p+=1
    q-=1
    p_not = p_not-2*q+2*p+1
  else:
    p+=1
    p_not = p_not+2*p+1

  x_cord.append(p)
  y_cord.append(q)



```

```python
print(x_cord)
```

```python
print(y_cord)
```

```python
for i in range(len(x_cord)):
  x = x_cord[i]
  y = y_cord[i]
  plt.scatter([x, -x, x, -x, y, -y, y, -y], [y, y, -y, -y, x, x, -x, -x], color='blue')
plt.show()
```
