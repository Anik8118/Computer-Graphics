```python
import matplotlib.pyplot as plt
```

```python
def xis(x1,x2,y1,y2):
  x_plot = []
  y_plot = []
  dx = x2-x1
  dy = y2-y1
  maximum = max(abs(dx), abs(dy))
  #increment_x = dx/maximum
  m = dy/dx

  for _ in range(dy if dy > dx else dx):
    if m <1:
      x1+=1
      x_plot.append(round(x1))
      y1+=1
      y_plot.append(round(y1))
    elif m>=1:
      x1+=1/m
      x_plot.append(round(x1))
      y1+=1
      y_plot.append(round(y1))
    else:
      x1+=1
      x_plot.append(round(x1))
      y1+=1
      y_plot.append(round(y1))

  plt.plot(x_plot,y_plot)
  plt.xlabel("X-Axis")
  plt.ylabel("Y-Axix")
  plt.title("DDA Algorithm")

  res = zip(x_plot, y_plot)
  for _ in res:
    print(_, end = " ")

  print(" ")

```

```python

x1 = int(input())
x2 = int(input())
y1 = int(input())
y2 = int(input())


xis(x1, x2, y1, y2)
```
